I appreciate several features of this graph. For one, if you look at the x-axis you can see that the first year they used in their data is written all the way out. However, the following years are abbreviated as to not make it too smushed together and hard to read. Similarly, the y-axis' top value, $10k, has the unit, $, and the k to represent 10,000 dollars. The following values, therefore, have no need for the $ or k and thus frees up more space. For easier comparison of the two sets of data they are trying to show, they simply overlapped the two grpahs. It isn't confusing though which graph matches which data because they have not only labeled each, but have also assigned a different color to each one. Due to the fact that the data stretches across the entire graph, they added gridlines to make it easier for the reader to see values at every point. To counteract the decrease in the data-ink ratio from adding the gridlines, they didn't include the frame. Finally, the title also does a good job of telling you what you are looking at.
